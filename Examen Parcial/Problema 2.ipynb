# Curva de Bézier y B-Spline

```python
def plotCurves(bez, bsp, points, name):
    x = var('x')
    a = plot([],figsize=(8, 8),title=name)
    classic_colors = ['red', 'green']
    for i in range(len(bez)): # Plots parametric polynomials
        p = bez[i]
        a += parametric_plot([p[0], p[1]], (x, p[2], p[2] + 1), color = 'red')
    for i in range(len(bsp) - 1): # Plots parametric polynomials
        p = bsp[i]
        a += parametric_plot([p[0], p[1]], (x, p[2], p[2] + bsp[-1]), color = 'green')
    for i in range(len(points) - 1): # Plots shape of points' polygon
        j = (i + 1) % len(points)
        p = [points[i][0] * x + points[j][0] * (1 - x), points[i][1] * x + points[j][1] * (1 - x),0]
        a += parametric_plot([p[0], p[1]], (x, p[2], p[2] + 1), color = 'blue')
    show(a)
```

```python
def Bezier(points): # 2D Bézier Curve for points in list "points"
    n = len(points) - 1
    # Preprocess binom(n, i) for all i = 0, ..., n O(n^2)
    # Uses property C(n, k) = C(n - 1, k - 1) + C(n - 1, k)
    # And memory saving for Dynamic Programming
    C = [0 for i in range(n + 1)]
    new_C = [0 for i in range(n + 1)]
    C[0] = C[1] = 1
    i = 2
    while i <= n:
        new_C[0] = new_C[i] = 1
        j = 1
        while j < i:
            new_C[j] = C[j - 1] + C[j]
            j += 1
        for j in range(i + 1):
            C[j] = new_C[j]
        i += 1
    # Preprocessing Bernstein Polynomials O(n^3)
    var('x')
    polys = [1 for i in range(n + 1)]
    prefix = 1
    for i in range(n + 1):
        polys[i] *= C[i] * prefix
        prefix *= x
    prefix = 1
    for i in range(n, -1, -1):
        polys[i] *= prefix
        prefix *= (1 - x)
    # Multiplying by each point and adding to parametric O(n^2)
    P = [0 for i in range(3)]
    for i in range(n + 1):
        for j in range(2):
            P[j] += polys[i] * points[i][j]
    return P
```

```python
def mergePolynomials(a, b):
    posA = 0
    posB = 0
    ans = []
    while posA < len(a) and posB < len(b):
        if a[posA][1] < b[posB][1]:
            ans.append(a[posA])
            posA += 1
        elif b[posB][1] < a[posA][1]:
            ans.append(b[posB])
            posB += 1
        else:
            ans.append([a[posA][0] + b[posB][0], a[posA][1]])
            posA += 1
            posB += 1
    while posA < len(a):
        ans.append(a[posA])
        posA += 1
    while posB < len(b):
        ans.append(b[posB])
        posB += 1
    return ans

def mergeParametric(a, b):
    posA = 0
    posB = 0
    ans = []
    while posA < len(a) and posB < len(b):
        if a[posA][2] < b[posB][2]:
            ans.append(a[posA])
            posA += 1
        elif b[posB][2] < a[posA][2]:
            ans.append(b[posB])
            posB += 1
        else:
            ans.append([a[posA][0] + b[posB][0], a[posA][1] + b[posB][1], a[posA][2]])
            posA += 1
            posB += 1
    while posA < len(a):
        ans.append(a[posA])
        posA += 1
    while posB < len(b):
        ans.append(b[posB])
        posB += 1
    return ans

def multiply(a, polys):
    cpoly = [[x[0] * a, x[1]] for x in polys]
    return cpoly

def multiplyPoint(a, polys):
    cpoly = [[x[0] * a[0], x[0] * a[1], x[1]] for x in polys]
    return cpoly
        
def BSpline(points, degK): # Uniform B-Spline of degree degK
    n = len(points) - 1
    m = n + degK + 1
    N = [ [ [] for i in range(degK + 1)] for j in range(m + 1)]
    t = [ i for i in range(m + 1)]
    var('x')
    for i in range(n + 1):
        N[i][1] = [ [1, t[i]] ] # B[i][k], N[i][k]
    for k in range(2, degK + 1):
        for i in range(n):
            A = multiply((x - t[i]) / (t[i + k - 1] - t[i]), N[i][k - 1])
            B = multiply((t[i + k] - x) / (t[i + k] - t[i+1]), N[i + 1][k - 1])
            C = mergePolynomials(A, B)
            N[i][k] = C
    polynomials = []
    result = []
    for i in range(n + 1):
        result = mergeParametric(result, multiplyPoint(points[i], N[i][degK]))
    result.pop()
    result.append(1)
    return result
```

```python
def plotBezierAndBSpline(points):
    Bez = Bezier(points)
    Bsp = BSpline(points, 3)
    plotCurves([Bez], Bsp, points, "Curvas de Bezier (Rojo) y B-Spline(verde)")
```
